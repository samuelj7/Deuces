```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "Resources/all_mens_matches.csv"

atp_2017_df = pd.read_csv(csv_path)
atp_2017_df
```

```python
country=atp_2017_df["Winner_Country"].value_counts()
country.head(20)
```

```python
clay_df=atp_2017_df.loc[atp_2017_df["Surface"]=="Clay"]
clay_df
winner_country=clay_df["Winner_Country"].value_counts()
winner_country=winner_country.rename(index="Clay")
winner_country.head(20)
```

```python
grass_df=atp_2017_df.loc[atp_2017_df["Surface"]=="Grass"]
grass_df
winner_country_grass=grass_df["Winner_Country"].value_counts()
winner_country_grass=winner_country_grass.rename(index="Grass")
winner_country_grass.head(20)
```

```python
hard_df=atp_2017_df.loc[atp_2017_df["Surface"]=="Hard"]
hard_df
winner_country_hard=hard_df["Winner_Country"].value_counts()
winner_country_hard=winner_country_hard.rename(index="Hard")
winner_country_hard.head(20)
```

```python
carpet_df=atp_2017_df.loc[atp_2017_df["Surface"]=="Carpet"]
carpet_df
winner_country_carpet=carpet_df["Winner_Country"].value_counts()
winner_country_carpet=winner_country_carpet.rename(index="Carpet")
winner_country_carpet.head(20)
```

```python
surface=atp_2017_df["Surface"].value_counts()
surface
```

```python
country_surface_df=atp_2017_df[["Tourney_Name","Surface","Winner_Country"]]
country_surface_df

new_df=atp_2017_df.groupby(country_surface_df["Winner_Country"])
new_df.head()
```

```python
clean_df = pd.DataFrame({
    "Clay":winner_country, 
    "Hard":winner_country_hard, 
    "Grass":winner_country_grass,
    "Carpet":winner_country_carpet})

clean_df_2=clean_df.fillna(0)
clean_df_2

sort_clean_df=clean_df_2.sort_values("Hard",ascending=False)
sort_clean_df
#sort_clean_df = clean_df_2.groupby(clean_df_2.index_values).sum().sort_values(ascending=False)
```

```python

plot_clean_df=sort_clean_df.head(20).plot(kind="bar",figsize=(100,20),fontsize=75)
plt.legend(prop={"size":50})
plt.savefig("Images/surface_wins_men_top_20.png")
plt.xticks(rotation=90)
```
