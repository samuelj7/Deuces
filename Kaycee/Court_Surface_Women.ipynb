```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "Resources/all_womens_matches.csv"

wta_df = pd.read_csv(csv_path)
wta_df
```

```python
wta_country=wta_df["Winner_Country"].value_counts()
wta_country.head(20)
```

```python
wta_clay_df=wta_df.loc[wta_df["Surface"]=="Clay"]
wta_clay_df
wta_winner_country_clay=wta_clay_df["Winner_Country"].value_counts()
wta_winner_country_clay=wta_winner_country_clay.rename(index="Clay")
wta_winner_country_clay.head(20)
```

```python
wta_grass_df=wta_df.loc[wta_df["Surface"]=="Grass"]
wta_grass_df
wta_winner_country_grass=wta_grass_df["Winner_Country"].value_counts()
wta_winner_country_grass=wta_winner_country_grass.rename(index="Grass")
wta_winner_country_grass.head(20)
```

```python
wta_hard_df=wta_df.loc[wta_df["Surface"]=="Hard"]
wta_hard_df
wta_winner_country_hard=wta_hard_df["Winner_Country"].value_counts()
wta_winner_country_hard=wta_winner_country_hard.rename(index="Hard")
wta_winner_country_hard.head(20)
```

```python
wta_carpet_df=wta_df.loc[wta_df["Surface"]=="Carpet"]
wta_carpet_df
wta_winner_country_carpet=wta_carpet_df["Winner_Country"].value_counts()
wta_winner_country_carpet=wta_winner_country_carpet.rename(index="Carpet")
wta_winner_country_carpet.head()
```

```python
wta_surface=wta_df["Surface"].value_counts()
wta_surface
```

```python
wta_country_surface_df=wta_df[["Tourney_Name","Surface","Winner_Country"]]
wta_country_surface_df

wta_new_df=wta_df.groupby(wta_country_surface_df["Winner_Country"])
wta_new_df.head()
```

```python
wta_clean_df = pd.DataFrame({
    "Clay":wta_winner_country_clay, 
    "Hard":wta_winner_country_hard, 
    "Grass":wta_winner_country_grass})

wta_clean_df_2=wta_clean_df.fillna(0)
wta_clean_df_2

sort_wta_clean_df=wta_clean_df_2.sort_values("Hard",ascending=False)
sort_wta_clean_df
```

```python
plot_wta_clean_df=sort_wta_clean_df.head(20).plot(kind="bar",figsize=(100,20),fontsize=75)
plt.legend(prop={"size":50})
plt.savefig("Images/surface_wins_women_top_20.png")
plt.xticks(rotation=90)
```
